```python
def snake_columns(data): # snake_case columns
    data.columns = [column.lower().replace(' ', '_') for column in data.columns]
    return data.sample(0)
```

<h3 style="color: #008080;">1. Load the dataset into a notebook.
</h3>

```python
data = pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
snake_columns(data)
```

<h3 style="color: #008080;">2. Create a barplot of the number of customers per <b>Gender</b>.</h3>

```python
data['gender'].value_counts().plot(kind='bar', xlabel='Gender', ylabel='number_of_customers', rot=0)
```

<h3 style="color: #008080;">3. Create a barplot of the number of customers per <b>EmploymentStatus</b> and <b>Gender</b>.
</h3>

```python
# Nara's idea with crosstab! I forgot how to plot I think...
crosstab = pd.crosstab(data.employmentstatus, data.gender)

# Create bar plot
plot = crosstab.plot.bar(rot=0, color=['#800080', '#FFB347'])

# Customize plot
plt.title('Gender Distribution by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Count')

# Show plot
plt.show()
```
